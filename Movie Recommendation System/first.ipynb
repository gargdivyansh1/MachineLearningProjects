```python
import numpy as np 
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('movies_data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
## selecting the relevant features for recommendations

selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
## replacing the null values with null string 

for feature in selected_features:
    data[feature] = data[feature].fillna('')
```

```python
## combining all the 5 selected features

comnbined_features = data['genres'] + ' ' + data['keywords'] + ' ' + data['tagline'] + ' ' + data['cast'] + ' ' + data['director']
```

```python
print(comnbined_features)
```

### converting the text data into the vector data

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(comnbined_features)
```

```python
print(feature_vector)
```

### Cosine Similarity

```python
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

### Getting the movie name from the user

```python
movie_name = input("Enter your favourite movie name : ")
```

```python
## creating the list for all the movies name given in the dataset 

movie_list = data['title'].tolist()
print(movie_list)
```

```python
## finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, movie_list)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
## finding the index of the movie based on the title of the movie

index = data[data.title == close_match]['index'].values[0]
print(index)
```

```python
## Similarity score 
similarity_score = list(enumerate(similarity[index]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
## sorting the movies on the basis of similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
## print the name of similar movies 

print("Movies suggested for you : \n")

i = 1 

for movie in sorted_similar_movies:
    index_of_movie = movie[0]
    title_from_index = data[data.index == index_of_movie]['title'].values[0]
    if i <= 30 :
        print(i, ':', title_from_index)
        i+=1
    else : break
```

### Movie recommendation System

```python
movie_name = input("Enter Your Favourite Movie Name : ")

list_of_all_movies = data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_movies)

close_match = find_close_match[0]

index_of_the_movie = data[data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similarity_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print("Movies Suggested For You Are : \n")

i = 1 

for movie in sorted_similarity_movies:
    index_of_movie = movie[0]
    title_for_movie = data[data.index == index_of_movie]['title'].values[0]
    if i <= 30 :
        print(i," : ", title_for_movie)
        i += 1
    else : break 
```
