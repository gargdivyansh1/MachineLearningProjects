```python
## converting the whole into a single fucntion for the betterment of the code

def reading_data(files):

    acc_df = pd.DataFrame()
    gyr_df = pd.DataFrame()

    acc_set = 1
    gyr_set = 1

    for f in files:
    
        participtant = f.split('-')[0].replace(data_path, '')
        label = f.split('-')[1]
        category = f.split('-')[2].rstrip('_MetaWear_2019').rstrip('12345')

        # print(participtant + ' ' + label + ' ' + category)

        df = pd.read_csv(f)

        df['participant'] = participtant
        df['label'] = label
        df['category'] = category

        if 'Accelerometer' in f:
            df['set'] = acc_set
            acc_set += 1
            acc_df = pd.concat([acc_df, df])

        if 'Gyroscope' in f:
            df['set'] = gyr_set
            gyr_set += 1
            gyr_df = pd.concat([gyr_df, df])

    acc_df.index = pd.to_datetime(acc_df['epoch (ms)'], unit= 'ms')
    gyr_df.index = pd.to_datetime(gyr_df['epoch (ms)'], unit='ms')

    del acc_df['epoch (ms)']
    del acc_df['time (01:00)']
    del acc_df['elapsed (s)']

    del gyr_df['epoch (ms)']
    del gyr_df['time (01:00)']
    del gyr_df['elapsed (s)']

    return acc_df, gyr_df
```

```python
acceleration_df, gyroscope_df = reading_data(files)
```

```python
## now as the data for both acceleration and gyroscope are founded then we could merge them toget the desired data frame 

# but in the each data frame there is a column for participant label and category which will hinder hence .. remove from any one of it 

data_merged = pd.concat([acceleration_df.iloc[:,:3], gyroscope_df], axis = 1)
```

```python
data_merged.head()
```

```python
# there is one more thing to notice which is that the frequency of the gyroscope is much more then the frequency of the acceleration .. it would be 100 percent sure that for some values of gyroscope there will be no value of the acceleration and for some there will be no value of the gyroscope ..

# hence dropping the null values rows 

data_merged.dropna()
```

```python
# Renaming the columns for better understanding

data_merged.columns = [
    'acc_x',
    'acc_y',
    'acc_z',
    'gyr_x',
    'gyr_y',
    'gyr_z',
    'participtant',
    'label',
    'category',
    'set'
]
```

```python
## now we need to reshape the data for the filling of the null values

sampling =  {
    'acc_x': 'mean',
    'acc_y': 'mean',
    'acc_z': 'mean',
    'gyr_x': 'mean',
    'gyr_y': 'mean',
    'gyr_z': 'mean',
    'participtant': 'last',
    'label': 'last',
    'category': 'last',
    'set': 'last'
}
data_merged[:1000].resample(rule='200ms').apply(sampling)
```

```python
## split by days
days = [g for n, g in data_merged.groupby(pd.Grouper(freq='D'))]
data_reshampled = pd.concat([df.resample(rule = '200ms').apply(sampling).dropna() for df in days]) # type: ignore

data_reshampled.info()
```
